```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic-passengers.csv', sep=";")
```

```python
df.head()
```

```python
dicSurvived = {"No":0,"Yes":1}
df["Survived"].replace(dicSurvived,inplace=True)
```

```python
df.head()
```

```python
dicSex = {"male":1, "female":0}
df["Sex"].replace(dicSex,inplace=True)
```

```python
df.head()
```

```python
df.drop(['Cabin'], axis=1)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model    import LinearRegression
from sklearn.metrics         import mean_squared_error
from sklearn                 import metrics
from sklearn.preprocessing   import PolynomialFeatures 
from sklearn.metrics         import mean_squared_error
from sklearn                 import metrics
from sklearn                 import svm, datasets
from sklearn.model_selection import GridSearchCV
```

```python
x_ = df[['Pclass', 'Sex', 'Age', 'Fare']].values
y_ = df['Survived'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_, y_, test_size=0.3, stratify=y_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sns
logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
logreg_proba = logreg.predict_proba(x_test) 
```

```python
logreg_proba = logreg_proba[:, 1] # to keep positive outcome
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
logreg_auc = roc_auc_score(y_test, logreg_proba)
logreg_auc 
```

```python
# Plotting the roc curve

logreg_x, logreg_y, _ = roc_curve(y_test, logreg_proba)
plt.plot(logreg_x, logreg_y)
plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```
